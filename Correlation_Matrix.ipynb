```python
#Importing required libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
data = pd.read_csv('Problem Statement Dataset.csv')
```

```python
#checking missing values
data.isnull().sum()
```

 Before calculating corrlation, drop the target variable
  because the correlation is concerned between independent variables 

```python
#dropping target variable
data = data.drop('Purchase', axis=1)
```

```python
#calculating correlation between independent variable
data.corr()
```

```python
#correlation matrix
Corr_Matrix = data.corr().abs()
```

```python
Corr_Matrix
```

```python
# selecting upper triangle of correlation matrix
upper = Corr_Matrix.where(np.triu(np.ones(Corr_Matrix.shape), k=1).astype(np.bool))
```

```python
upper
```

```python
# finding index of variables with correlation greater than 0.5
to_drop = [column for column in upper.columns if any(upper[column] > 0.5)]
```

```python
to_drop
```

* 'Product_Category_2', 'Product_Category_3' are the most correlated values

```python
#dropping the variables and craeting new dataset
new_data = data.drop(data[to_drop], axis=1)
```
